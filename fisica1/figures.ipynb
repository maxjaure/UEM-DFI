```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi,100); y = 1+np.sin(x);
```

```python
fig, ax = plt.subplots()
ax.set_xlabel(r"$t$(s)", fontsize = "20")
ax.set_ylabel(r"$x$(m)", fontsize = "20")
ax.tick_params(direction = "in", labelsize = "20")
ax.grid("on")
ax.plot(x,y)
fig.savefig("xtcurve.pdf",bbox_inches="tight")
```
