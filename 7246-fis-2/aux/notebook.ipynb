```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 6*np.pi, 600); y = np.cos(x); z = np.sin(x);
```

```python
fig, ax = plt.subplots(figsize=(5,3))
# ax.set_title(r'$\mathrm{Gr√°fico\, do}\, \sin x$',fontsize=20)
ax.set_xticks([0,np.pi,2*np.pi,3*np.pi,4*np.pi,5*np.pi,6*np.pi])
ax.grid(True)
ax.set_xticklabels([r'$0$',r'$\pi$',r'$2\pi$',r'$3\pi$',r'$4\pi$',r'$5\pi$',r'$6\pi$'])
ax.tick_params(direction='in',labelsize=14)
ax.set_xlabel(r'$x$',fontsize=16)
ax.set_ylabel(r'$y$',fontsize=16)
ax.plot(x,y,label=r'$\cos\, x$')
ax.plot(x,z,label=r'$\mathrm{sen}\, x$')
ax.legend(fontsize=16)
```

```python
fig.savefig("senocosseno.pdf",bbox_inches='tight')
```

```python
t = np.linspace(0, 20, 200); y = np.exp(-0.1*t)*np.cos(np.sqrt(1-0.2**2/4)*t); z = np.exp(-0.1*t); y1 = np.exp(-t); y2 = (np.exp((-3+np.sqrt(8))*t)+np.exp((-3-np.sqrt(8))*t))/2
```

```python
fig, ax = plt.subplots(figsize=(7,5))
# ax.set_title(r'$\mathrm{Gr√°fico\, do}\, \sin x$',fontsize=20)
ax.grid(b=True)
ax.tick_params(direction='in',labelsize=14)
ax.set_xlabel(r'$t$',fontsize=16)
ax.set_ylabel(r'$x$',fontsize=16)
ax.plot(t,y,label=r'$b=0.2\sqrt{km}$')
ax.plot(t,z,'--')
ax.plot(t,y1,label=r'$b=2\sqrt{km}$')
ax.plot(t,y2,label=r'$b=6\sqrt{km}$')
ax.legend(fontsize=16)
```

```python
fig.savefig("osciladoramortecido.pdf",bbox_inches='tight')
```
